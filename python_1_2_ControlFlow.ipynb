# The 'print()' function. 
The 'print()' function displays output as text and is useful for showing results of calculations as well as for debugging code. Text should be within  "" and numeric values when being shown concatenated with text need to be cast to string.

```python
# Here is an example of the print statement
print("Hello World")
print("The sum of 2 and 3 is ",2+3)
print("The sum of 2 and 3 is " + str(2+3))
print("The sum of 2 and 3 is =",str(2+3))
print("The sum of 2 and 3 is", "= " + str(2+3) +" units")
```

# The 'if' Statement
The 'if, elif, else' statement is a conditional statement and executes specific pieces of code depending on the condition evaluating to True or False.The 'if' statement exits after the executing the first statement that evaluates to being True.

```python
#Here is an example
if (5>3):
    print("5 greater than 3")
else:
    print("5 is less than or equal to 3")

if (7<5):
    print("7 is greater or equal than 5")
else:
    print("7 is less than 5")
    
```

# The 'while' loop
The 'while' loop repeats a piece of code as long as the 'while' condition holds True. It exits  when the condition is False.
Care must be taken while writing a 'while' loop that it has a terminating condition else it may loop infinitely.

```python
# Here is an example of 'while' loop which moves Ghotul on the grid as long as front is clear
while Ghotul.front_is_clear():
    Ghotul.move()
```

# The Range function
The range function generates a sequence of numbers that can be used in loops or as an index. It takes upto 3 arguments 
, the starting number, the ending number and the increment step size.
The Range sequence takes values upto but not including the ending number.

```python
#Here is an example of the Range function
print (range(4))
print (range(1,4,1))
print (range(-1,4,1))
```

# The 'for' loop
The 'for' loop is similar to the 'while' loop , the main difference being that the number of iterations of the loop needs to be known in advance. It is a safer option when the number of times we want the loop to execute is already known. This prevents a situation of 'infinite loop' from arising.

```python
# Here is an example of 'for' loop which moves Ghotul 4 steps in the direction it is facing
#Here i takes each value from the sequence one at a time i.e 0,1,2,3 in this case. Note when not specified, the sequence starts from 0
for i in range(4):
    Ghotul.move()
```
