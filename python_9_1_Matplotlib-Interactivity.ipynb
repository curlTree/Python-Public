# Matplotlib - Animation
using the Function FuncAnimation from the library matplotlib.animation for animating charts 


```python
#this line is needed to show animation
import matplotlib.animation as s
import matplotlib.pyplot as plt
import numpy as np
n=100
X=np.random.randn(n)
def update(curr):
    
    if (curr==100): #the update functions which keeps running till iteration value is reached 
        a.event_source.stop()#code to halt the iteration on reaching 100
    plt.cla()  #clear the figure
    print(curr)
    bins=np.arange(-4,4,0.5) #create class intervals of width 0.5 ,the random x values falling in each bin will increase the height of that bar
#     X=np.random.randn(curr) #This will refresh the entire series
    plt.hist(X[:curr],bins=bins) #This progressively builds up the histogram graph, everytime recreating all the points already covered
    plt.axis([-4,4,0,30]) # set the x-axis limits and increments
    plt.gca().set_title('Animated Random distribution')
    plt.gca().set_ylabel('ylabel')
    plt.gca().set_xlabel('xlabel')
    plt.annotate('n={}'.format(curr),[3,27]) # test to appear on the chart,denoting the iteration number
fig=plt.figure()    #draw the frame of the figure
# the Animation function which calls the 'update' function 100 times 
a=s.FuncAnimation(fig,update,interval=100)    
```

```python
# interactivity in charts
%matplotlib notebook
import matplotlib.pyplot as plt
import numpy as np

plt.figure()
data=np.random.rand(10)
plt.plot(data)

def onclick(event):
    
    plt.cla()
    plt.plot(data)
    plt.plot(round(event.x/56,2),round(event.y/355,2),'rx')
    plt.text(round(event.x,2),round(event.y,2),'Event at ({},{})'.format(round(event.x,2),round(event.y,2)))
    print('inside')
    plt.gca().set_title('Event at {},{},{} and data {},{}'.format(round(event.x,2),round(event.y,2),'\n',round(event.xdata,2)
                                                                  ,round(event.ydata,2)))
    # note the format function required for displaying the {} in the title
    
#click on a point anywhere on the chart to display the point and the data associated with it
plt.gcf().canvas.mpl_connect('button_press_event',onclick);
    
```

```python
from random import shuffle
import pandas as pd
origins = ['China', 'Brazil', 'India', 'USA', 'Canada', 'UK', 'Germany', 'Iraq', 'Chile', 'Mexico']

shuffle(origins)

df = pd.DataFrame({'height': np.random.rand(10),
                   'weight': np.random.rand(10),
                   'origin': origins})
```

```python
plt.figure()
df
# picker=5 means the mouse doesn't have to click directly on an event, but can be up to 5 pixels away
plt.scatter(df['height'], df['weight'], picker=5)
plt.gca().set_ylabel('weight')
plt.gca().set_xlabel('height')
```

```python
def onpick(event):
    print(event.ind)
    origin=df.iloc[event.ind[0]]['origin']
    plt.gca().set_title('Item came from {},{}'.format(origin,event.ind))
plt.gcf().canvas.mpl_connect('pick_event', onpick) #tied to the gcf() and child mpl_connect 
#click on any point on the scatter plot to display the country/region which it belongs to
```
