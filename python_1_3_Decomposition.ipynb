# Functions
Functions enable decomposition of code into sub tasks that usually repeat . This makes the code easier to handle and debug. A function defintion starts with a keyword 'def' , a function name and optionally arguments or parameters that can be passed to the function.A function can be called from anywhere in the code by specifying the function name along with the arguments that it takes.Indentation is important while writing a function definition.The argument variables and any other variables declared in the function are local in scope to the function i.e. they do not retain their values outside of the function. A function may return a value to the variable from which it is being called.

```python
# Here is an example of a function definition and call
def multiply(x,y):
    return x*y

print("Product of 7 and 5 is",multiply(7,5))
```
