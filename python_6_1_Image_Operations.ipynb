# Image Handling
Images are also arrays of pixel values,typically of the format (height,width,3), the 3 stands for the primary colours Red,Green,Blue.Each pixel can take a value between 0 to 255 for each of the primary colours e.g. a pixel with values (255,0,0) is a dark Red pixel, while (0,255,0) is a dark Green pixel , (255,0,255) is a deep Yellow pixel.
Hence images can be edited and used just like numeric arrays.Note it requires the graphing and image library 'matplotlib' to be imported


```python
# Here are some examples of  image file operations
import matplotlib.pyplot as plt

img_array=plt.imread('test_image.jpg','wb')
#copy the array into another array as the original array is read-only
img_copy=img_array.copy()
plt.imshow(img_array)
plt.show()
print(img_array.shape)
```

```python
# an example of editing an image based on pixel colour thresholds
# read the height and width and number of colours of the image array
i_width,i_height,i_channel=img_array.shape
for w in range(i_width): # go through each row of the image
    for h in range(i_height): # for each row go through each column
         #apply a red color filter i.e. if Red component is >200 then make it 200
        if img_copy[w,h,0]>200: #img_array[h,w,0] refers to Red,img_array[h,w,1] refers to Green,img_array[h,w,2] refers to Blue
            img_copy[w,h,0]=200
plt.imshow(img_copy)  
plt.show()
```
